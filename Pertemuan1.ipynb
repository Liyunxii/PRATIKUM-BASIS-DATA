**ðŸ“˜ Pokok Bahasan.**

Topik utama mencakup aturan-aturan untuk mentransformasikan berbagai komponen ERD.

*   **Entitas Kuat (Strong Entity):** Setiap entitas kuat menjadi satu tabel. Atribut kunci (key attribute) menjadi Primary Key.
*   **Atribut (Attributes):** Atribut Komposit: Hanya komponen-komponen dari atribut komposit yang menjadi kolom, bukan atribut itu sendiri.

*   **Atribut Multivalue:** Atribut multivalue harus menjadi tabel tersendiri yang baru.

*   **Atribut Derivatif:** Atribut derivatif (turunan) tetap menjadi kolom dalam tabel.

*   **Entitas Lemah (Weak Entity):**
Setiap entitas lemah menjadi tabel baru. Atribut kunci dari entitas kuat yang terhubung akan menjadi foreign key (FK) di tabel entitas lemah.
*   **Relasi (Relations):**


1.   **Relasi 1:1 (Satu ke Satu):**
Menghasilkan dua tabel. Atribut kunci dari salah satu entitas menjadi foreign key di tabel entitas lainnya.
2.   **Relasi 1:N (Satu ke Banyak):**
Menghasilkan dua tabel. Atribut kunci dari sisi 'satu' (1) menjadi foreign key di tabel sisi 'banyak' (N).
3. **Relasi M:N (Banyak ke Banyak):**
Menghasilkan tiga tabel. Relasi itu sendiri menjadi tabel baru (tabel asosiatif). Atribut kunci dari kedua entitas yang terhubung menjadi foreign key di tabel relasi baru tersebut.
4. **Relasi Unary:**
Relasi ke entitas itu sendiri (1 tabel). Menambahkan foreign key baru di tabel yang sama yang mengacu pada primary key-nya sendiri.
5. **Relasi Ternary (Derajat Tiga):**
Menghasilkan empat tabel. Satu tabel untuk relasi, dan tiga tabel untuk entitas yang terlibat.

**Konsep Lanjutan:**

Generalisasi/Spesialisasi (GENSPEC):

Dibahas dua metode;

(1) Superclass dan semua subclass menjadi tabel, atau

(2) Hanya subclass yang menjadi tabel, dengan mewarisi atribut superclass.

**Agregasi (Aggregation):**

Relasi antara entitas yang beragregasi akan memiliki foreign key dari semua entitas yang berhubungan.

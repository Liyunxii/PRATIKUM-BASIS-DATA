MySQL menyediakan beberapa tipe data yang umum digunakan:


```python
import pandas as pd

data = {
    'Tipe Data': ['Int', 'Float', 'Date', 'DateTime', 'Char', 'VarChar', 'Blob', 'LongBlob'],
    'Keterangan': ['Angka', 'Angka Desimal', 'Tanggal', 'Tanggal & Waktu', 'String', 'String', 'String', 'String'],
    'Range/Format': ['-2147483648 – 2147483648', '[Tidak disebutkan]', 'YYYY-MM-DD', 'YYYY-MM-DD HH:MM:SS', '1 – 255 Char', '1 – 255 Char', '<= 65535 Char', '<= 4294967295 Char']
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Mengatur ulang indeks agar dimulai dari 1
# range(1, len(df) + 1) menghasilkan urutan 1, 2, 3, 4, 5, 6, 7, 8
df.index = range(1, len(df) + 1)

# Menampilkan tabelnya
df
```

**2.10 Data Definition Languange (DDL).**

**DDL** adalah perintah yang digunakan untuk mendefinisikan struktur database.




```python
import pandas as pd

data = {
    'Perintah DDL': ['Membuat Database', 'Melihat seluruh Database', 'Mengakses Database', 'Menghapus Database'],
    'Fungsi': ['Untuk membuat database baru di server MySQL.', 'Untuk melihat semua database yang telah dibuat.', 'Untuk memilih dan menggunakan database.', 'Untuk menghapus (drop) database yang telah dibuat.'],
    'Sintaks': ['create database nama_database;', 'show databases;', 'use nama_database;', 'drop database nama_database;'],
    'Contoh': ['create database praktikum;', '[Tidak ada contoh di modul]', 'use praktikum;', 'drop database nama_database;']
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Mengatur ulang indeks agar dimulai dari 1
# range(1, len(df) + 1) menghasilkan urutan 1, 2, 3, 4
df.index = range(1, len(df) + 1)

# Menampilkan tabelnya
df
```
